```python
#VERSI 1

import random
import numpy as np

# Inisialisasi data
order_sepatu = [
    {'order_id': 1, 'ukuran': '38', 'bahan': 'Kulit'},
    {'order_id': 2, 'ukuran': '40', 'bahan': 'Kanvas'},
    {'order_id': 3, 'ukuran': '39', 'bahan': 'Kulit'},
    # Tambahkan order sepatu lainnya sesuai kebutuhan
]

bahan_baku = {
    'Kulit': 100,
    'Kanvas': 80,
    # Tambahkan jenis bahan baku lainnya sesuai kebutuhan
}

# Definisikan parameter algoritma genetika
populasi_size = 50
generasi = 100
probabilitas_crossover = 0.8
probabilitas_mutasi = 0.2

# Fungsi fitness
def fitness(individu):
    total_kekurangan_bahan = 0
    for order in individu:
        bahan = order['bahan']
        if bahan_baku[bahan] <= 0:
            total_kekurangan_bahan += 1
        else:
            bahan_baku[bahan] -= 1
    return 1 / (1 + total_kekurangan_bahan)

# Fungsi seleksi menggunakan metode turnamen
def seleksi(populasi, fitness_populasi):
    k = 5
    selected = []
    for _ in range(len(populasi)):
        candidates = random.choices(populasi, k=k, weights=fitness_populasi)
        selected.append(max(candidates, key=lambda x: fitness(x)))
    return selected

# Fungsi crossover menggunakan one-point crossover
def crossover(parent1, parent2):
    point = random.randint(1, len(parent1) - 1)
    child1 = parent1[:point] + parent2[point:]
    child2 = parent2[:point] + parent1[point:]
    return child1, child2

# Fungsi mutasi dengan menukar posisi dua order sepatu
def mutasi(individu):
    idx1, idx2 = random.sample(range(len(individu)), 2)
    individu[idx1], individu[idx2] = individu[idx2], individu[idx1]
    return individu

# Inisialisasi populasi awal
populasi = []
for _ in range(populasi_size):
    individu = order_sepatu.copy()
    random.shuffle(individu)
    populasi.append(individu)

# Proses algoritma genetika
for _ in range(generasi):
    fitness_populasi = [fitness(individu) for individu in populasi]

    # Seleksi
    selected_populasi = seleksi(populasi, fitness_populasi)

    # Crossover
    offspring_populasi = []
    for i in range(0, populasi_size, 2):
        parent1 = selected_populasi[i]
        parent2 = selected_populasi[i+1]
        if random.random() < probabilitas_crossover:
            child1, child2 = crossover(parent1, parent2)
            offspring_populasi.append(child1)
            offspring_populasi.append(child2)
        else:
            offspring_populasi.append(parent1)
            offspring_populasi.append(parent2)

    # Mutasi
    for i in range(populasi_size):
        if random.random() < probabilitas_mutasi:
            offspring_populasi[i] = mutasi(offspring_populasi[i])

    # Ganti populasi dengan offspring
    populasi = offspring_populasi

# Cari individu dengan fitness tertinggi
fitness_populasi = [fitness(individu) for individu in populasi]
index_terbaik = np.argmax(fitness_populasi)
solusi_terbaik = populasi[index_terbaik]

# Output hasil penjadwalan order produksi
print("Solusi Terbaik:")
for order in solusi_terbaik:
    print("Order ID:", order['order_id'], "- Ukuran:", order['ukuran'], "- Bahan:", order['bahan'])
```

```python
#versi 13 menampilkan semua data order_sepatu pada output hasil penjadwalan order produksi:
import random
import numpy as np

# Inisialisasi data
order_sepatu = [
    {'order_id': 1, 'ukuran': '38', 'bahan': 'Kulit', 'jumlah': 1},
    {'order_id': 2, 'ukuran': '40', 'bahan': 'Kanvas', 'jumlah': 2},
    {'order_id': 3, 'ukuran': '39', 'bahan': 'Kulit', 'jumlah': 1},
    # Tambahkan order sepatu lainnya sesuai kebutuhan
]

bahan_baku = {
    'Kulit': 100,
    'Kanvas': 80,
    # Tambahkan jenis bahan baku lainnya sesuai kebutuhan
}

jumlah_order = len(order_sepatu)
jumlah_mesin = 3

# Definisikan parameter algoritma genetika
populasi_size = 50
generasi = 100
probabilitas_crossover = 0.9
probabilitas_mutasi = 0.1

# Fungsi fitness
def fitness(individu):
    bahan_baku_tersedia = bahan_baku.copy()
    total_kekurangan_bahan = 0
    for order in individu:
        bahan = order['bahan']
        ukuran = order['ukuran']
        jumlah = order['jumlah']
        if bahan_baku_tersedia[bahan] < jumlah * int(ukuran):
            total_kekurangan_bahan += 1
        else:
            bahan_baku_tersedia[bahan] -= jumlah * int(ukuran)
    return 1 / (1 + total_kekurangan_bahan)

# Fungsi seleksi menggunakan metode turnamen
def seleksi(populasi, fitness_populasi):
    k = 5
    selected = []
    for _ in range(len(populasi)):
        candidates = random.choices(populasi, k=k, weights=fitness_populasi)
        selected.append(max(candidates, key=lambda x: fitness(x)))
    return selected

# Fungsi crossover menggunakan one-point crossover
def crossover(parent1, parent2):
    point = random.randint(1, len(parent1) - 1)
    child1 = parent1[:point] + parent2[point:]
    child2 = parent2[:point] + parent1[point:]
    return child1, child2

# Fungsi mutasi dengan menukar posisi dua order sepatu
def mutasi(individu):
    idx1, idx2 = random.sample(range(len(individu)), 2)
    individu[idx1], individu[idx2] = individu[idx2], individu[idx1]
    return individu

# Inisialisasi mesin
mesin = ['Mesin A', 'Mesin B', 'Mesin C']  # Ubah sesuai dengan jumlah mesin yang ada

# Inisialisasi populasi awal
populasi = []
for _ in range(populasi_size):
    individu = []
    for order in order_sepatu:
        for _ in range(order['jumlah']):
            individu.append(order.copy())
    random.shuffle(individu)
    populasi.append(individu)

# Proses algoritma genetika
for _ in range(generasi):
    fitness_populasi = [fitness(individu) for individu in populasi]

    # Seleksi
    selected_populasi = seleksi(populasi, fitness_populasi)

    # Crossover
    offspring_populasi = []
    for i in range(0, populasi_size, 2):
        parent1 = selected_populasi[i]
        parent2 = selected_populasi[i+1]
        if random.random() < probabilitas_crossover:
            child1, child2 = crossover(parent1, parent2)
            offspring_populasi.append(child1)
            offspring_populasi.append(child2)
        else:
            offspring_populasi.append(parent1)
            offspring_populasi.append(parent2)

    # Mutasi
    for i in range(populasi_size):
        if random.random() < probabilitas_mutasi:
            offspring_populasi[i] = mutasi(offspring_populasi[i])

    # Ganti populasi dengan offspring
    populasi = offspring_populasi

# Cari individu dengan fitness tertinggi
fitness_populasi = [fitness(individu) for individu in populasi]
index_terbaik = np.argmax(fitness_populasi)
solusi_terbaik = populasi[index_terbaik]

# Update ketersediaan bahan baku berdasarkan solusi terbaik
bahan_baku_terbaik = bahan_baku.copy()
for order in solusi_terbaik:
    bahan = order['bahan']
    jumlah = order['jumlah']
    ukuran = order['ukuran']
    bahan_baku_terbaik[bahan] -= jumlah * int(ukuran)

# Output hasil penjadwalan order produksi
print("Output hasil penjadwalan order produksi:")
for order in order_sepatu:
    order_id = order['order_id']
    ukuran = order['ukuran']
    bahan = order['bahan']
    jumlah = order['jumlah']
    ketersediaan_bahan = bahan_baku_terbaik[bahan]
    mesin_index = order_sepatu.index(order) % jumlah_mesin
    mesin_terpakai = mesin[mesin_index]
    for _ in range(jumlah):
      if ketersediaan_bahan == -bahan_baku_terbaik[bahan]:
        ketersediaan_bahan = 0
      print("Order ID:", order_id, "- Ukuran:", ukuran, "- Bahan:", bahan, "- Jumlah:", jumlah, "- Ketersediaan Bahan:", ketersediaan_bahan, "- Mesin:", mesin_terpakai)
```
